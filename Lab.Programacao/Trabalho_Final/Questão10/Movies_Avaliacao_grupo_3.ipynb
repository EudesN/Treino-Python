##Gráfico que representa a distribuição de arrecadação dos 20 filmes com maior bilheteria mundial

primeiro, calculamos os valores da arrecadação e guardaremos em uma lista

```python
filmes_maior_bilheteria = df_movies_limpo.sort_values('worldwide_gross', ascending=False).head(20)

# Calcula o lucro de cada filme (bilheteria mundial - custo de produção)
arrecadacao = filmes_maior_bilheteria['worldwide_gross']

# Converte a Series para lista diretamente
dados_y = arrecadacao.tolist()
```

agora utilizaremos a lista para construir o gráfico

```python
titulos = filmes_maior_bilheteria['title']

plt.figure(figsize=(14, 8))
barras = plt.bar(titulos, dados_y, color='skyblue')

for i in range(len(barras)):
    altura = barras[i].get_height()
    plt.text(
        barras[i].get_x() + barras[i].get_width() / 2,
        altura / 2,             # posição vertical no meio da barra
        f'{dados_y[i]:,.0f}',  # valor formatado
        ha='center',
        va='center',
        fontsize=9,
        color='black',
        rotation=90             # texto na vertical
    )

plt.title('Arrecadação Estimada dos 20 Filmes com Maior Bilheteria Mundial', fontsize=16)
plt.ylabel('Arrecadação Estimada', fontsize=12)
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```

##Gráfico que agrupa os filmes por gênero e mostra o percentual de participação da bilheteria mundial de todos os filmes do dataset por gênero

```python
arrecadacao_por_genero = df_movies_limpo.groupby('genre')['worldwide_gross'].sum()
percentuais = arrecadacao_por_genero / arrecadacao_por_genero.sum() * 100
percentuais
```

```python
percentuais_ordenado = percentuais.sort_values(ascending=True)
# Criar gráfico de pizza
plt.figure(figsize=(10, 10))
plt.pie(percentuais_ordenado, labels=percentuais_ordenado.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 10})

plt.title('Participação Percentual da Bilheteria Mundial por Gênero', fontsize=14)
plt.axis('equal')  # mantém o formato circular
plt.tight_layout()
plt.show()
```
